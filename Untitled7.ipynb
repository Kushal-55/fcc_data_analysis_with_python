```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("/content/drive/MyDrive/medical_examination.csv")
df.head()
```

```python
'''
Add an overweight column to the data. To determine if a person is overweight, first calculate their BMI by dividing their weight in kilograms by the square of their height in meters.
If that value is > 25 then the person is overweight.
Use the value 0 for NOT overweight and the value 1 for overweight.
'''
df['overweight']= (df['weight']/(df['height']/100)**2)
df['overweight'] = (df['overweight']>25).astype(int)
df.head()
```

```python
'''
Normalize the data by making 0 always good and 1 always bad.
If the value of cholesterol or gluc is 1, make the value 0. If the value is more than 1, make the value 1.
'''
df.cholesterol.replace(1,0)
df.loc[df['cholesterol']>1 ,'cholesterol'] =0
df.gluc.replace(1,0)
df.loc[df.gluc >1,'gluc'] = 0
```

```python
'''
Convert the data into long format and create a chart that shows the value counts of the categorical features using seaborn's catplot(). 
The dataset should be split by 'Cardio' so there is one chart for each cardio value.
'''
df2 = pd.melt(df, id_vars=['cardio'], value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight'])
df2 = df2.groupby(['cardio', 'variable', 'value'])['value'].count().to_frame()
df2.rename(columns={'value': 'total'}, inplace=True)
df2.reset_index(inplace=True)
plot = sns.catplot(x='variable', y='total', hue='value', col='cardio', kind='bar', data=df2)
figure = plot.fig
```

```python
'''
Clean the data. Filter out the following patient segments that represent incorrect data:
diastolic pressure is higher than systolic (Keep the correct data with (df['ap_lo'] <= df['ap_hi']))
height is less than the 2.5th percentile (Keep the correct data with (df['height'] >= df['height'].quantile(0.025)))
height is more than the 97.5th percentile
weight is less than the 2.5th percentile
weight is more than the 97.5th percentile
'''
h_25 = df['height'].quantile(0.025)
h_75 = df['height'].quantile(0.975)
w_25 = df['weight'].quantile(0.025)
w_75 = df['weight'].quantile(0.975)
df_heat = df[df['ap_lo'] <= df['ap_hi']]
df_heat = df_heat[df_heat['height'] >= h_25]
df_heat = df_heat[df_heat['height'] <= h_75]
df_heat = df_heat[df_heat['weight'] >= w_25]
df_heat = df_heat[df_heat['weight'] <= w_75]

# heat plot
correlation = df_heat.corr()
correlation = correlation.round(1)
mask = np.triu(np.ones_like(correlation, dtype=np.bool))

fig, ax = plt.subplots(figsize=(12, 9))
heat_map = sns.heatmap(correlation, mask=mask, annot=True)
heat_map.set_yticklabels(heat_map.get_yticklabels(), rotation=0)
heat_map.set_xticklabels(heat_map.get_xticklabels(), rotation=90)
```
